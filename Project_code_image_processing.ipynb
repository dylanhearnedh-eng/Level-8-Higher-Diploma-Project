This code is used to download the image files from my Google Drive and convert them into a CSV for analysis

```python
#Connects Colab to the Google drive
from google.colab import drive
drive.mount('/content/drive')
```

```python
import os
import numpy as np
import pandas as pd
from PIL import Image, ImageOps

data_dir = "/content/drive/MyDrive/Project Images"   # Folder containing all labelled subfolders

rows = [] #creates an empty list

#For looks that cycle through the images checking folder and file type
for label in sorted(os.listdir(data_dir)):
    folder = os.path.join(data_dir, label)
    if not os.path.isdir(folder):
        continue

    for filename in os.listdir(folder):
        if not filename.lower().endswith((".jpg", ".png")):
            continue

        img_path = os.path.join(folder, filename)

        # Load all images
        img = Image.open(img_path).convert("L")   #makes image greyscale, "L" means 8-bit greyscale
        img = img.resize((32, 32))                #resizes the image, these images already at this size
        img = ImageOps.autocontrast(img)          #increases contrast of image for better accuracy%

        pixels = np.array(img).flatten()          #flattens array

        rows.append(list(pixels) + [label])

df = pd.DataFrame(rows)
df.to_csv("images_with_labels.csv", index=False) #names the new CSV file

print("job done") #just confirms that the file is ready
```
